b) Define a cost function
$J(Î¸)$
that will allow you to compute the cost on the training and
cross-validation sets.


$$J(\theta) = \frac{1}{2N} \sum_{n=1}^{N} (h_{\theta}(\mathbf{x}_{n}) - y_{n})^{2}$$

where $h_{\theta}(\mathbf{x}_{n})$ is your prediction.

```python
def compute_cost(y_pred, y_true):
    N = y_pred.shape[0]
    return (1.0 / (2.0 * N)) * np.sum((y_pred - y_true)**2)
```

```python
plt.figure(figsize=(12,6))
plt.grid(alpha=0.5)
plt.scatter(X_train, y_train, label="Train")
plt.scatter(X_test, y_test,label="Test")

train_cost_curve = []
test_cost_curve = []
for order in range(10):
    
    # Test on train set
    thetas = fit(X_train, y_train, order)
    y_pred = predict(thetas, X_train)
    cost_train = compute_cost(y_pred, y_train)
    train_cost_curve.append(cost_train)

    # Test on test set
    # thetas = fit(X_test, y_test, order)
    y_pred = predict(thetas, X_test)
    cost_test = compute_cost(y_pred, y_test)
    test_cost_curve.append(cost_test)

    X_new = np.linspace(-1, 11, 100)
    plt.plot(X_new, predict(thetas, X_new))

    print(f"Order={order}; cost on train set: {cost_train:.2f}; cost on test set: {cost_test:.2f}")

plt.legend([f"Order {i}" for i in range(10)]);
```

**h) Explain how under/over-fitting is involved here.**

The bigger the order, the more the model will fit closely to the points in the dataset, with poorer generalisation because of sticking too close, following too much the variance, and failing to capture the bigger data patterns thus --> overfitting
In the opposite having a too little order can result in a too simple model with high biais, failing to understand the complexity of the data patterns
